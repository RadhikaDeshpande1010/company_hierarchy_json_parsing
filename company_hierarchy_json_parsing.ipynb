# Company Hierarchy JSON Parsing
## Demo: Load and Display Nested JSON Data

```python
import json

# Load JSON file
with open("/content/sample_data/company_nested_data.json", "r") as json_file:
    json_data = json.load(json_file)

# Extract top-level fields
company_details   = json_data["company"]
company_name      = company_details["name"]
department_list   = company_details["departments"]
company_locations = company_details["locations"]

# Display company overview
print("========== COMPANY DETAILS ==========")
print("Company Name:", company_name)

print("\n========== DEPARTMENT DETAILS ==========")
for department in department_list:
    department_id   = department["dept_id"]
    department_name = department["dept_name"]
    employee_list   = department["employees"]

    print("\n" + "-" * 30)
    print("Dept Id:", department_id)
    print("Department Name:", department_name)

    for employee in employee_list:
        employee_id     = employee["emp_id"]
        employee_name   = employee["name"]
        employee_skills = employee["skills"]
        employee_salary = employee["salary"]

        print("\nEmployee Id:", employee_id)
        print("Employee Name:", employee_name)
        print("Employee Skills:", employee_skills)
        print("Employee Salary:", employee_salary)

print("\n========== COMPANY LOCATIONS ==========")
print("Company Locations:", company_locations)
```

## Question 1: Company Name

```python
'''
Question 1: Company Name

Task:
- Read the JSON data and print the company name.

Expected Output:
- Company Name: TechNova
'''

# Extract company details
company_details = json_data["company"]

# Extract and print company name
company_name = company_details["name"]
print("Company Name:", company_name)
```

## Question 2: Total Departments

```python
'''
Question 2: Total Departments

Task:
- Find the total number of departments in the company.

Expected Output:
- Total Departments: 3
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Count total departments
department_count = 0
for department in department_list:
    department_count += 1

print("Total Departments:", department_count)
```

## Question 3: Departments with Python Skill

```python
'''
Question 3: Departments with Python Skill

Task:
- Find all departments that have at least one employee with Python skill.

Expected Output:
- IT
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# List to store department names with at least one Python-skilled employee
python_skill_departments = []

for department in department_list:
    department_name = department["dept_name"]
    employee_list   = department["employees"]

    for employee in employee_list:
        employee_skills = employee["skills"]

        if "Python" in employee_skills:
            python_skill_departments.append(department_name)
            break  # No need to check further employees in this department

print("Departments having at least one employee with Python skill:")
for department_name in python_skill_departments:
    print(department_name)
```

## Question 4: Total Employees

```python
'''
Question 4: Total Employees

Task:
- Count the total number of employees across all departments.

Expected Output:
- Total Employees across all departments: 7
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Sum employees across all departments
total_employee_count = 0
for department in department_list:
    employee_list         = department["employees"]
    total_employee_count += len(employee_list)

print("Total Employees across all departments:", total_employee_count)
```

## Question 5: Employees in IT Department

```python
'''
Question 5: Employees in IT Department

Task:
- Print the ID and name of every employee in the IT department.

Expected Output:
- ID: 1  |  Name: Amit
- ID: 2  |  Name: Riya
- ID: 3  |  Name: Suresh
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store IT department employees (employee_id : employee_name)
it_department_employees = {}

for department in department_list:
    department_name = department["dept_name"]
    employee_list   = department["employees"]

    if department_name == "IT":
        for employee in employee_list:
            employee_id   = employee["emp_id"]
            employee_name = employee["name"]
            it_department_employees[employee_id] = employee_name

print("Employees in IT Department:")
for employee_id, employee_name in it_department_employees.items():
    print(f"  ID: {employee_id}  |  Name: {employee_name}")
```

## Question 6: Employees with Python Skill

```python
'''
Question 6: Employees with Python Skill

Task:
- Find and print names of all employees who have Python as a skill.

Expected Output:
- Amit
- Suresh
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# List to store names of Python-skilled employees
python_skilled_employees = []

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name   = employee["name"]
        employee_skills = employee["skills"]

        if "Python" in employee_skills:
            python_skilled_employees.append(employee_name)

print("Employees with Python Skill:")
for employee_name in python_skilled_employees:
    print(employee_name)
```

## Question 7: Average Salary of All Employees

```python
'''
Question 7: Average Salary of All Employees

Task:
- Calculate the average salary of all employees.

Expected Output:
- Average Salary of All Employees: 57571.43
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Initialize counters
total_employee_count = 0
total_salary_sum     = 0

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        total_employee_count += 1
        total_salary_sum     += employee["salary"]

# Calculate average after both loops complete
average_salary = total_salary_sum / total_employee_count

print(f"Average Salary of All Employees: {average_salary:.2f}")
```

## Question 8: Highest Paid Employee

```python
'''
Question 8: Highest Paid Employee

Task:
- Find the name and salary of the highest paid employee.

Expected Output:
- Karan is the highest paid employee with a salary of 70000.
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Initialize highest salary tracker
highest_salary             = 0
highest_paid_employee_name = ""

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name   = employee["name"]
        employee_salary = employee["salary"]

        if employee_salary > highest_salary:
            highest_salary             = employee_salary
            highest_paid_employee_name = employee_name

print(f"{highest_paid_employee_name} is the highest paid employee with a salary of {highest_salary}.")
```

## Question 9: Department-wise Employee Count

```python
'''
Question 9: Department-wise Employee Count

Task:
- Create a dictionary with department name as key
  and number of employees as value.

Expected Output:
- {"IT": 3, "HR": 2, "Finance": 2}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store department-wise employee count
department_employee_count = {}

for department in department_list:
    department_name = department["dept_name"]
    employee_list   = department["employees"]

    for employee in employee_list:
        if department_name in department_employee_count:
            department_employee_count[department_name] += 1
        else:
            department_employee_count[department_name] = 1

print("Department-wise Employee Count:")
print(department_employee_count)
```

## Question 10: Company Locations

```python
'''
Question 10: Company Locations

Task:
- Print all company office locations, each on a separate line.

Expected Output:
- Delhi
- Mumbai
- Bangalore
- Hyderabad
'''

# Extract company details and locations
company_details   = json_data["company"]
company_locations = company_details["locations"]

print("Company Locations:")
for location in company_locations:
    print(location)
```

## Question 11: Employee Name → Department Mapping

```python
'''
Question 11: Employee Name → Department Mapping

Task:
- Create a dictionary where:
    key   = employee name
    value = department name

Expected Output:
- {"Amit": "IT", "Riya": "IT", ...}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee name and their department name
employee_department_mapping = {}

for department in department_list:
    department_name = department["dept_name"]
    employee_list   = department["employees"]

    for employee in employee_list:
        employee_name = employee["name"]
        employee_department_mapping[employee_name] = department_name

print("Employee Name → Department Mapping:")
print(employee_department_mapping)
```

## Question 12: Employee ID → Uppercase Name

```python
'''
Question 12: Employee ID → Uppercase Name

Task:
- Create a dictionary where:
    key   = employee ID
    value = employee name in UPPERCASE

Expected Output:
- {1: "AMIT", 2: "RIYA", ...}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee ID and uppercase employee name
employee_id_uppercase_name_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_id   = employee["emp_id"]
        employee_name = employee["name"]
        employee_id_uppercase_name_mapping[employee_id] = employee_name.upper()

print("Employee ID → Uppercase Name Mapping:")
print(employee_id_uppercase_name_mapping)
```

## Question 13: Employee Name → First Character

```python
'''
Question 13: Employee Name → First Character

Task:
- Create a dictionary where:
    key   = employee name
    value = first character of the name

Expected Output:
- {"Amit": "A", "Riya": "R", ...}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee name and its first character
employee_name_first_char_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name = employee["name"]
        employee_name_first_char_mapping[employee_name] = employee_name[0].upper()

print("Employee Name → First Character Mapping:")
print(employee_name_first_char_mapping)
```

## Question 14: Employee Name → Name Length

```python
'''
Question 14: Employee Name → Name Length

Task:
- Create a dictionary where:
    key   = employee name
    value = length of the employee name

Expected Output:
- {"Amit": 4, "Suresh": 6, ...}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee name and its character length
employee_name_length_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name        = employee["name"]
        employee_name_length = len(employee_name)
        employee_name_length_mapping[employee_name] = employee_name_length

print("Employee Name → Name Length Mapping:")
print(employee_name_length_mapping)
```

## Question 15: Department Name → Uppercase

```python
'''
Question 15: Department Name → Uppercase

Task:
- Create a dictionary where:
    key   = department name
    value = department name in UPPERCASE

Expected Output:
- {"IT": "IT", "HR": "HR", "Finance": "FINANCE"}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store department name and its uppercase version
department_uppercase_mapping = {}

for department in department_list:
    department_name = department["dept_name"]
    department_uppercase_mapping[department_name] = department_name.upper()

print("Department Name → Uppercase Mapping:")
print(department_uppercase_mapping)
```

## Question 16: Employee Name → Skills String

```python
'''
Question 16: Employee Name → Skills String

Task:
- Create a dictionary where:
    key   = employee name
    value = all skills joined into a single comma-separated string

Expected Output:
- {"Amit": "Python, SQL", "Karan": "Accounting, Excel"}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee name and list of their skills
employee_skill_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name   = employee["name"]
        employee_skills = employee["skills"]

        if employee_name in employee_skill_mapping:
            employee_skill_mapping[employee_name].extend(employee_skills)
        else:
            employee_skill_mapping[employee_name] = list(employee_skills)

# Convert skill lists to comma-separated strings
employee_skill_string_mapping = {
    employee_name: ", ".join(skills)
    for employee_name, skills in employee_skill_mapping.items()
}

print("Employee Name → Skills String Mapping:")
print(employee_skill_string_mapping)
```

## Question 17: Skill → Employee Names String

```python
'''
Question 17: Skill → Employee Names String

Task:
- Create a dictionary where:
    key   = skill name
    value = employee names separated by "|"

Expected Output:
- {"Python": "Amit|Suresh", "Excel": "Rahul|Karan"}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store skill and list of employees having that skill
skill_employee_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name   = employee["name"]
        employee_skills = employee["skills"]

        for skill in employee_skills:
            if skill in skill_employee_mapping:
                skill_employee_mapping[skill].append(employee_name)
            else:
                skill_employee_mapping[skill] = [employee_name]

# Convert employee name lists to "|"-separated strings
skill_employee_string_mapping = {
    skill: "|".join(employee_names)
    for skill, employee_names in skill_employee_mapping.items()
}

print("Skill → Employee Names Mapping:")
print(skill_employee_string_mapping)
```

## Question 18: Employee Name → Last Character

```python
'''
Question 18: Employee Name → Last Character

Task:
- Create a dictionary where:
    key   = employee name
    value = last character of the name

Expected Output:
- {"Amit": "t", "Pooja": "a"}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee name and last character of their name
employee_name_last_char_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name = employee["name"]
        employee_name_last_char_mapping[employee_name] = employee_name[-1]

print("Employee Name → Last Character Mapping:")
print(employee_name_last_char_mapping)
```

## Question 19: Department Name → Employee Name Lengths

```python
'''
Question 19: Department Name → Employee Name Lengths

Task:
- Create a dictionary where:
    key   = department name
    value = list of lengths of employee names in that department

Expected Output:
- {"IT": [4, 4, 6], "HR": [4, 5]}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store department name and list of employee name lengths
department_employee_name_lengths = {}

for department in department_list:
    department_name = department["dept_name"]
    employee_list   = department["employees"]

    for employee in employee_list:
        employee_name        = employee["name"]
        employee_name_length = len(employee_name)

        if department_name in department_employee_name_lengths:
            department_employee_name_lengths[department_name].append(employee_name_length)
        else:
            department_employee_name_lengths[department_name] = [employee_name_length]

print("Department Name → Employee Name Length List:")
print(department_employee_name_lengths)
```

## Question 20: Employee Name → Reversed Name

```python
'''
Question 20: Employee Name → Reversed Name

Task:
- Create a dictionary where:
    key   = employee name
    value = reversed spelling of the employee name

Expected Output:
- {"Amit": "timA", "Riya": "ayiR"}
'''

# Extract company and department details
company_details = json_data["company"]
department_list = company_details["departments"]

# Dictionary to store employee name and its reversed version
employee_name_reversed_mapping = {}

for department in department_list:
    employee_list = department["employees"]

    for employee in employee_list:
        employee_name          = employee["name"]
        reversed_employee_name = employee_name[::-1]
        employee_name_reversed_mapping[employee_name] = reversed_employee_name

print("Employee Name → Reversed Name Mapping:")
print(employee_name_reversed_mapping)
```
